### Ветвления. Анализ попадания точки с заданной координатой в область


```python
x, y = float(input()), float(input())
r = float(input())
if x ** 2 + y ** 2 > r ** 2:
    print('in')
elif x ** 2 + y ** 2 < r ** 2:
    print('out')
else:
    print('bound')
```

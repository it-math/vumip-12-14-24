#### Выбор элемента по индексу. 
Индекс элемента указывается в квадратных скобках, нумерация в Python начинается с нуля:

```python
text = "Hello world. We love python"
print(text[0])
```

Отрицательные индексы в Python тоже существуют – в таком случае элементы отсчитываются с конца последовательности:

```python
print(text[-1])
```

Стандартный способ (не все языки поддерживают нумерацию с конца) извлечения последнего элемента на основе общей длины последовательности тоже работает, но считается избыточным:

```python
print(text[len(text)-1])
```

```python
s = 'hello' + 'python'
print(s, len(s))
```

### Срезы

```python
s = 'abcdefgh'
ss = s[2:5]
print(ss)
ss1 = s[1:7:2]
print(ss1)
ss2 = s[3:]
print(ss2)
ss3 = s[:6]
print(ss3)
ss4 = s[::-1]
print(ss4)
ss5 = s[1:7:-1]
print(ss5)
ss6 = s[-1:-7:-2]
print(ss6)
s1 = s[::2]
s2 = s[1::2]
#s[0] = 'A'
s = 'A' + s[1:]
```

### Экранированные последовательности

```python
s = 'ab\ncd\tvn\vabv'
print(s)
```

```python
s = 'C:\\number\\n.txt'
s1 = r'C:\number\n.txt'
print(s1)
```

#### Функции и методы строк

```python
s = 'abcdefabaaaf'
k = s.find('a')
print(k)
```

```python
s = 'abcdefabaaaf'
k1 = s.rfind('a')
print(k1)
```

```python
s = 'abcdefabaaaf'
k2 = s.find('x')
print(k2)
```

```python
s = 'abcdefabaaaf'
k3 = s.find('a', 3, 9)
#k3 = s.find(str,[start],[end])
print(k3)
```

```python
s = 'abcdefabaaaf'
kk = s.find('bcd')
print(kk)
```

```python
s = 'abcdefabaaaf'
k = s.index('x')
print(k)
```

```python
s = 'agaaghgaaaghaaa'
s = s.replace('a', 'A', 3)
print(s)
```

```python
s = 'agaaghgaaaaaghaaa'
s = s.replace('aa', 'a a')
print(s)
```

```python
k = s.count('a')
print(k)
```

```python
s = 'agaaghgaaaaaghaaa'
k = s.count('aa')
print(k)
```

```python
txt = "Hello my FRIENDS"
x = txt.lower()
print(x)
```

```python
txt = "Hello my friends"
x = txt.upper()
print(x)
```

```python
txt = "Welcome to my world"
x = txt.title()
print(x)
```

```python
txt = "Hello My Name Is PETER"
x = txt.swapcase()
print(x)
```

```python
s = 'привет, мир!'
s = s.capitalize()
print(s)
```

```python
txt = "banana"
x = txt.ljust(20)
print(x, "is my favorite fruit.")
```

```python
txt = "banana"
x = txt.rjust(20)
print(x, "is my favorite fruit.")
```

```python
txt = "     banana     "
x = txt.strip()
print("of all fruits", x, "is my favorite")
```

```python
txt = "I could eat bananas all day"
x = txt.partition("bananas")
print(x)
```

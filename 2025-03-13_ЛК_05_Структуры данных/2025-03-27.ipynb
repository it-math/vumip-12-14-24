```python
x = 0.7
s = 1
eps = 0.000001
i = 1
an = (-1)**i * x ** (2*i) / 2 / i
while abs(an) > eps:
    s += an
    i += 1
    an = (-1)**i * x ** (2*i) / 2 / i
    if i > 100:
        print('ряд не сходится')
        break
print(s, i-1)
```

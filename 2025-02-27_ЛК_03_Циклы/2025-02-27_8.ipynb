```python
s = 0
x = 2
for i in range(0, 11):
    for j in range(1, 6):
        s += x ** (i + j) / (2 * i + 2 * j)
print(s)
```

```python
s = 0
x = 2
for i in range(0, 11):
    j = 1
    while j <= 5:
        s += x ** (i + j) / (2 * i + 2 * j)
        j += 1
print(s)
```
